## Filled nan values in total_episode

```python
merged_df.loc[merged_df["url"].str.contains("/film"), "total_episode"] = merged_df.loc[merged_df["url"].str.contains("/film"), "total_episode"].fillna(1)
```

```python
merged_df[merged_df["total_episode"].isna()]
```

```python
merged_df.loc[merged_df["name"] == "After the Show Ends", "total_episode"] = merged_df.loc[merged_df["name"] == "After the Show Ends", "total_episode"].fillna(8)
merged_df.loc[merged_df["name"] == "Bir Doksan (One Ninety)	", "total_episode"] = merged_df.loc[merged_df["name"] == "Bir Doksan (One Ninety)	", "total_episode"].fillna(5)
merged_df.loc[merged_df["name"] == "Smile", "total_episode"] = merged_df.loc[merged_df["name"] == "Smile", "total_episode"].fillna(24)
merged_df.loc[merged_df["name"] == "Kill It	", "total_episode"] = merged_df.loc[merged_df["name"] == "Kill It	", "total_episode"].fillna(12)
```

```python
merged_df["total_episode"] = merged_df["total_episode"].astype("Int64")
```

```python
merged_df.info()
```

## create new column by movie or serie

```python
def determine_type(url):
    if "/film" in url:
        return "Movie"
    else:
        return "Series"

merged_df["type"] = merged_df["url"].apply(determine_type)
```

## Cast column was split 

```python
data = []

for i, entry in enumerate(merged_df["cast"]):
    if not pd.isna(entry):
        for cast_dict in ast.literal_eval(entry):
            name = cast_dict['name']
            character = cast_dict['caracter']
            series_name = merged_df.loc[i, "name"]  
            likes = merged_df.loc[i, "like"]
            watch = merged_df.loc[i, "watch"]
            data.append({"Name": name, "Character": character, "Series": series_name, "Likes": likes, "watch": watch})

cast_df = pd.DataFrame(data)

cast_df
```

# Top 10 of the TV series the actors starred in, according to their total like

```python
popularity = cast_df.groupby('Name')['Likes'].sum().reset_index().sort_values(by='Likes', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y='Name', x='Likes', data=popularity, palette="viridis")
plt.title('The most Popular players')
plt.xlabel('Sum of likes')
plt.ylabel('player name')
plt.show()
```

# Number of TV series played by actors

```python
actor_counts = cast_df['Name'].value_counts()

top_actors = actor_counts[actor_counts > 14]

#These 4 blocks of code are related to creating colors. because some actors have played in the same number of plays
unique_counts = top_actors.unique()
color_palette = sns.color_palette("hsv", len(unique_counts))
color_dict = {count: color_palette[i] for i, count in enumerate(unique_counts)}
colors = [color_dict[count] for count in top_actors]

plt.figure(figsize=(10, 6))
top_actors.plot.pie(autopct='%1.1f%%', startangle=140, colors = colors)
plt.title('Top 10 Actors by Number of Series')
plt.ylabel('')  # Y label'ı gizle
plt.show()
```

```python
lee_min_hoo_shows = cast_df[cast_df['Name'].str.contains('Lee Min-Ho', case=False, na=False)]
lee_min_hoo_ratings = merged_df[merged_df['name'].isin(lee_min_hoo_shows['Series'])][['name', 'watch']]

plt.figure(figsize=(12, 6))
sns.barplot(x='watch', y='name', data=lee_min_hoo_ratings, palette="viridis")
plt.title('Lee Min-Hoo\'s series ')
plt.xlabel('Watch')
plt.ylabel('Series Name')
plt.show()
```

# We splited the category list and counted how many there were

```python
# Kategorileri ayıklamak ve veriyi genişletmek için bir DataFrame oluşturuyoruz
data = []

for i, row in merged_df.iterrows():
    categories = row['category'].split(',')
    for category in categories:
        category = category.strip().lower()
        data.append({'name': row['name'], 'category': category, 'rating': row['rating'], "year":row["year"]})
                     
category_df = pd.DataFrame(data)
```

```python
category_df.info()
```

```python
categories = category_df['category'].unique()

plt.figure(figsize=(15, 40))
plt.subplots_adjust(hspace=0.5)
for i, category in enumerate(categories, 1):
    plt.subplot(len(categories)//2 + 1, 2, i)
    sns.histplot(category_df[category_df['category'] == category]['rating'], bins=20)
    plt.title(f'{category.capitalize()} Kategorisi İçin Puan Dağılımı')
    plt.xlabel('Puan')
    plt.ylabel('Dizi Sayısı')

plt.show()
```

```python
from collections import Counter
categories_list = []

for cat_string in merged_df["category"]:
    categories = cat_string.split(",")
    
    for category in categories:
        category = category.strip().lower()
        
        categories_list.append(category)

category_counts = Counter(categories_list)
merged_df["category_count"] = merged_df["category"].apply(lambda x: len(x.split(",")))
```
