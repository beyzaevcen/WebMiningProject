### Data Frame for AI

```python
df_for_ai = merged_df.copy()
```

```python
# Kategorileri içeren bir liste oluşturun
categories_list = []

# "category" sütununu işleyin
for cat_string in df_for_ai["category"]:
    # Virgülle ayrılmış kategorileri parçalayın
    categories = cat_string.split(",")
    
    # Her bir kategori için
    for category in categories:
        # Boşlukları temizleyin
        category = category.strip().lower()
        
        # Kategoriyi listeye ekleyin
        categories_list.append(category)

# Her bir kategori için yeni bir sütun oluşturun
for category in set(categories_list):
    df_for_ai[category] = df_for_ai["category"].str.contains(category, case=False).astype(int)
```

```python
df_for_ai.drop("category",axis=1 ,inplace=True)
```

```python
df_for_ai["type"] = df_for_ai["type"].map({"Series": 1, "Movie": 0})
```

```python
df_for_ai.drop("url",axis=1 ,inplace=True)
```
